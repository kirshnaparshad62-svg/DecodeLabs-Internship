### Outlier Detection Using IQR

```python
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["TotalPrice"] < lower) |
    (df["TotalPrice"] > upper)
]

print("Outliers:", len(outliers))
```

### Remove Outliers

```python
df = df[
    (df["TotalPrice"] >= lower) &
    (df["TotalPrice"] <= upper)
]

print(df.shape)
```

### Detect Outlier Using Z-Score

```python
df["ZScore"] = zscore(df["TotalPrice"])

outliers = df[df["ZScore"].abs() > 3]

print("Z-score Outliers:", len(outliers))
```

**Remove:**

```python
df = df[df["ZScore"].abs() <= 3]
```

**Drop helper column:**

```python
df.drop("ZScore", axis=1, inplace=True)
```

### Feature Engineering

**Feature 1: Delivery Days**

```python
df["DeliveryDays"] = (
    df["DeliveryDate"] - df["OrderDate"]
).dt.days

OrderDate = "2023-01-01"
DeliveryDate = "2023-01-05"

DeliveryDays = 4
```

**Feature 2: Revenue After Discount**

```python
df["RevenueAfterDiscount"] = (
    df["Quantity"] *
    df["UnitPrice"] *
    (1 - df["Discount"])
)
```

**Feature 3: Profit Estimate**

```python
df["EstimatedProfit"] = (
    df["RevenueAfterDiscount"] -
    df["ShippingCost"]
)
```

**Feature 4: Sales Category**

```python
df["SalesCategory"] = pd.cut(
    df["TotalPrice"],
    bins=[0,1000,3000,6000,12000],
    labels=["Low","Medium","High","Premium"]
)
```

### Encode Categorical Variables

```python
df = pd.get_dummies(
    df,
    drop_first=True
)
```

### Feature Scaling

```python
scaler = StandardScaler()

numeric_cols = df.select_dtypes(
    include=["int64","float64"]
).columns

df[numeric_cols] = scaler.fit_transform(
    df[numeric_cols]
)
```

### Save Clean Dataset

```python
df.to_csv(
    "Cleaned_Product_Sales.csv",
    index=False
)

print("Dataset Saved Successfully")
```
